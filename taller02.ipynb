### Ejercicio 1: La Tienda

Se desea crear una aplicación que permita manejar el inventario de una tienda, conocer cuánto dinero hay en caja y tener un control de estadística de ventas.

La tienda maneja cuatro productos. De cada una de ellos se maneja la siguiente información:

* Nombre: no puede haber dos productos con el mismo nombre.
* Tipo: un string, puede ser uno de los siguientes: papelería, supermercado o droguería.
* Cantidad actual del producto en tienda: número de unidades disponibles para la venta que en la bodega.
* Cantidad mínima para abastecimiento: número de productos por debajo del cual se puede hacer un nuevo pedido al proveedor.
* Precio base de venta por unidad o valor unitario del producto.

Para calcular el precio final de cada producto, se debe sumar los impuestos que defina la ley (el IVA). Dichos impuestos dependen del tipo del producto, de la siguiente manera:

- Papelería: 16%
- Supermercado: 4%
- Droguería: 12%

Eso quiere decir que si un lápiz tiene un precio base de $10, el precio final será de $11,6 considerando que un lápiz es un producto de papelería, y sobre estos se debe pagar el 16% de impuestos.

### Pruebas de la clase Producto

```python
# Estas son las pruebas de la clase producto
import unittest

# Casos de prueba
class PruebaProducto(unittest.TestCase):

    def setUp(self):
        """Configuración de los objetos que vamos a usar en las pruebas"""
        self.lapiz = Producto("lapiz", "papeleria", 550, 18, 5)
        self.aspirina = Producto("aspirina", "drogueria", 109.5, 25, 8)
        self.pan = Producto("pan", "supermercado", 150.0, 15, 20)

    def test_constructor_getters(self):
        """Prueba del constructor de la clase producto"""
        self.prod = Producto("cuaderno", "papeleria", 300.0, 5, 2)
        self.assertEqual(self.prod.dar_nombre(), "cuaderno")
        self.assertEqual(self.prod.dar_tipo(), "papeleria")
        self.assertEqual(self.prod.dar_cantidad_bodega(), 5)
        self.assertEqual(self.prod.dar_cantidad_minima(), 2)
        self.assertEqual(self.prod.dar_valor_unitario(), 300.0)
        self.assertEqual(self.prod.dar_cantidad_unidades_vendidas(), 0)

    def test_dar_precio_final(self):
        """Prueba del método de dar el precio final de un producto"""
        lapiz = self.lapiz
        aspirina = self.aspirina
        pan = self.pan
        # -----
        self.assertAlmostEqual(638.0, lapiz.dar_precio_final(), places=1)
        self.assertAlmostEqual(122.64, aspirina.dar_precio_final(), places=2)
        self.assertAlmostEqual(156.0, pan.dar_precio_final(), places=1)
    

    def test_vender(self):
        """Prueba del método vender de la clase producto"""
        self.assertEqual(18, self.lapiz.dar_cantidad_bodega())
        resultado = self.lapiz.vender(10)
        self.assertEqual(8, self.lapiz.dar_cantidad_bodega())
        self.assertEqual(10, resultado)
        self.assertEqual(638.0, self.lapiz.dar_precio_final())
        self.assertEqual(10, self.lapiz.dar_cantidad_unidades_vendidas())

        resultado = self.lapiz.vender(30)
        self.assertEqual(0, self.lapiz.dar_cantidad_bodega())
        self.assertEqual(8, resultado)
        self.assertEqual(638.0, self.lapiz.dar_precio_final())
        self.assertEqual(18, self.lapiz.dar_cantidad_unidades_vendidas())

        self.assertEqual(15, self.pan.dar_cantidad_bodega())
        resultado = self.pan.vender(5)
        self.assertEqual(10, self.pan.dar_cantidad_bodega())
        self.assertEqual(5, self.pan.dar_cantidad_unidades_vendidas())
        resultado = self.pan.vender(5)
        self.assertEqual(5, self.pan.dar_cantidad_bodega())
        self.assertEqual(10, self.pan.dar_cantidad_unidades_vendidas())


    def test_abastecer(self):
        """Prueba del método abastecer de la clase Producto"""
        self.assertEqual(18, self.lapiz.dar_cantidad_bodega())
        resultado = self.lapiz.abastecer(15)
        self.assertEqual(33, self.lapiz.dar_cantidad_bodega())
        self.assertEqual(5, self.lapiz.dar_cantidad_minima())
        self.assertEqual(33, resultado)

        resultado = self.lapiz.abastecer(-10)
        self.assertEqual(33, resultado)
        self.assertEqual(33, self.lapiz.dar_cantidad_bodega())

        resultado = self.lapiz.abastecer(17)
        self.assertEqual(50, resultado)
        self.assertEqual(50, self.lapiz.dar_cantidad_bodega())


    def test_puede_abastecer(self):
        """Prueba del método puede_abastecer de la clase producto"""
        self.assertFalse(self.lapiz.puede_abastecer())
        self.assertFalse(self.aspirina.puede_abastecer())
        self.assertTrue(self.pan.puede_abastecer())


# ----------------------------------------
# Aquí se ejecutan los casos de prueba
unittest.main(argv=['', '-v', 'PruebaProducto'], verbosity=2, exit=False)
```

### Pruebas de la clase Tienda

```python
# Pruebas de la clase Tienda
import unittest

# Los casos de prueba
class PruebaTienda(unittest.TestCase):

    def setUp(self):
        """Configuración de los objetos que vamos a usar en las pruebas"""
        self.lapiz = Producto("lapiz", "papeleria", 550, 18, 5)
        self.aspirina = Producto("aspirina", "drogueria", 109.5, 25, 8)
        self.pan = Producto("pan", "supermercado", 150.0, 15, 20)
        self.borrador = Producto("borrador", "papeleria", 207.3, 30, 10)
        self.tienda = Tienda(self.lapiz, self.aspirina, self.borrador, self.pan, 0.0)

    # Primera prueba
    def test_constructor_tienda(self):
        self.assertEqual(self.lapiz, self.tienda.dar_producto1())
        self.assertEqual(self.aspirina, self.tienda.dar_producto2())
        self.assertEqual(self.borrador, self.tienda.dar_producto3())
        self.assertEqual(self.pan, self.tienda.dar_producto4())
        self.assertEqual(0.0, self.tienda.dar_dinero_en_caja())

    def test_dar_producto_por_nombre(self):
        self.assertEqual(self.lapiz, self.tienda.dar_producto_por_nombre("lapiz"))
        self.assertEqual(self.pan, self.tienda.dar_producto_por_nombre("pan"))
        self.assertEqual(self.borrador, self.tienda.dar_producto_por_nombre("borrador"))
        self.assertIsNone(self.tienda.dar_producto_por_nombre("queso"))

    def test_dar_cantidad_total_producto(self):
        """Prueba del método de la cantidad total de productos"""
        self.assertEqual(88, self.tienda.dar_cantidad_total_productos())
        self.lapiz.vender(10)
        self.assertEqual(78, self.tienda.dar_cantidad_total_productos())

    def test_dar_promedio_cantidad_vendidas(self):
        self.lapiz.vender(5)
        self.assertEqual(13, self.lapiz.dar_cantidad_bodega())
        self.aspirina.vender(10)
        self.assertEqual(15, self.aspirina.dar_cantidad_bodega())
        self.borrador.vender(5)
        self.assertEqual(25, self.borrador.dar_cantidad_bodega())
        self.pan.vender(5)
        self.assertEqual(10, self.pan.dar_cantidad_bodega())
        self.assertEqual(6.25, self.tienda.dar_promedio_cantidad_vendidas())

    def test_dar_promedio_precio_final(self):
        self.assertEqual(289.277, self.tienda.dar_promedio_precio_final())

    def test_dar_producto_mas_vendido(self):
        """Prueba del método del producto más vendido"""
        self.lapiz.vender(5)
        self.assertEqual(13, self.lapiz.dar_cantidad_bodega())
        self.aspirina.vender(10)
        self.assertEqual(15, self.aspirina.dar_cantidad_bodega())
        self.borrador.vender(5)
        self.assertEqual(25, self.borrador.dar_cantidad_bodega())
        self.pan.vender(5)
        self.assertEqual(10, self.pan.dar_cantidad_bodega())
        self.assertEqual(self.aspirina, self.tienda.dar_producto_mas_vendido())
        self.borrador.vender(12)
        self.assertEqual(13, self.borrador.dar_cantidad_bodega())
        self.assertEqual(17, self.borrador.dar_cantidad_unidades_vendidas())
        self.assertEqual(self.borrador, self.tienda.dar_producto_mas_vendido())

    def test_dar_producto_menor_cantidad_bodega(self):
        self.assertEqual(self.pan, self.tienda.dar_producto_menor_cantidad_bodega())
        self.pan.abastecer(20)
        self.assertEqual(35, self.pan.dar_cantidad_bodega())
        self.assertEqual(self.lapiz, self.tienda.dar_producto_menor_cantidad_bodega())

    def test_dar_nombre_producto_mas_barato(self):
        nom = self.tienda.dar_nombre_producto_mas_barato()
        self.assertEqual("aspirina", nom)
        t = Tienda(Producto("arroz", "supermercado", 250, 10, 1), \
                   Producto("cuaderno", "papeleria", 65, 5, 2), \
                   Producto("iguaraya", "supermercado", 34, 11, 6), \
                   Producto("crema no. 4", "drogueria", 7500, 100, 10), 3400)
        self.assertEqual("iguaraya", t.dar_nombre_producto_mas_barato())

    def test_dar_cantidad_productos_nombre_inicia_vocal(self):
        self.assertEqual(1, self.tienda.dar_cantidad_productos_nombre_ini_vocal())
        t = Tienda(Producto("arroz", "supermercado", 250, 10, 1), \
            Producto("cuaderno", "papeleria", 65, 5, 2), \
            Producto("iguaraya", "supermercado", 34, 11, 6), \
            Producto("crema no. 4", "drogueria", 7500, 100, 10), 3400)
        self.assertEqual(2, t.dar_cantidad_productos_nombre_ini_vocal())

    def test_dar_cantidad_productos_pueden_abastecer(self):
        self.assertEqual(1, self.tienda.dar_cantidad_productos_pueden_abastecer())
        self.borrador.vender(25)
        self.assertEqual(5, self.borrador.dar_cantidad_bodega())
        self.assertEqual(2, self.tienda.dar_cantidad_productos_pueden_abastecer())
        t = Tienda(Producto("arroz", "supermercado", 250, 10, 1), \
            Producto("cuaderno", "papeleria", 65, 5, 2), \
            Producto("iguaraya", "supermercado", 34, 11, 6), \
            Producto("crema no. 4", "drogueria", 7500, 100, 10), 3400)
        self.assertEqual(0, t.dar_cantidad_productos_pueden_abastecer())
        t = Tienda(Producto("arroz", "supermercado", 250, 10, 100), \
            Producto("cuaderno", "papeleria", 65, 5, 200), \
            Producto("iguaraya", "supermercado", 34, 11, 600), \
            Producto("crema no. 4", "drogueria", 7500, 100, 1000), 3400)
        self.assertEqual(4, t.dar_cantidad_productos_pueden_abastecer())

    def test_vender(self):
        res = self.tienda.vender_producto("lapiz", 10)
        self.assertAlmostEqual(6380.0, self.tienda.dar_dinero_en_caja(), places=1)
        self.assertEqual(10, res)
        self.assertEqual(8, self.lapiz.dar_cantidad_bodega())
        res = self.tienda.vender_producto("borrador", 40)
        self.assertAlmostEqual(13594.04, self.tienda.dar_dinero_en_caja(), places=2)
        self.assertEqual(30, res)
        self.assertEqual(0, self.borrador.dar_cantidad_bodega())

        
    def test_abastecer(self):
        res = self.tienda.abastecer_producto("lapiz", 5)
        self.assertEqual(23, res)
        self.assertEqual(23, self.lapiz.dar_cantidad_bodega())
        res = self.tienda.abastecer_producto("borrador", -10)
        self.assertEqual(30, self.borrador.dar_cantidad_bodega())


# -----------------
unittest.main(argv=['', '-v', 'PruebaTienda'], verbosity=2, exit=False)
```
